```python
import streamlit as st
import numpy as np
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from PIL import Image
import os
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import time
import plotly.graph_objects as go
from wordcloud import WordCloud






df = pd.read_csv('Job.csv')

all_skills = df['required skills'].dropna().str.split(',')  # Split by commas
flat_skills = [item.strip() for sublist in all_skills for item in sublist if item.strip()]  # Flatten and clean

# 2. Create a dictionary with word counts (frequency of each skill)
word_freq = pd.Series(flat_skills).value_counts().to_dict()

# 3. Generate a word cloud using the skill frequency
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)

# 4. Display the word cloud
st.write("#### Basic Skills Required")
plt.figure(figsize=(8, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide the axis
st.pyplot(plt)
```
